```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Crear un DataFrame con los datos proporcionados
data = pd.DataFrame({
    'Persona': range(1, 11),
    'Altura': [173, 100, 50, 180, 210, 160, 155, 176, 162, 210],
    'Peso': [85, 40, 30, 90, 130, 70, 68, 86, 71, 120]
})

# Divisi√≥n en variables predictora (X) y variable de salida (y)
X = data[['Altura']]
y = data['Peso']
```

```python
# Divisi√≥n en conjunto de entrenamiento y prueba (70-30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Modelo de regresi√≥n lineal
model = LinearRegression()
model.fit(X_train, y_train)
# Predicciones
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
print(f"El b0 o intercepto es: {model.intercept_}")
print(f"El b1 o coeficiente: {model.coef_}")

# Visualizaci√≥n de los errores
plt.scatter(X_test, y_test, color='blue', label='Datos de prueba')
plt.plot(X_test, y_test_pred, color='red', label='L√≠nea de regresi√≥n')
for i in range(len(X_test)):
    plt.plot([X_test.iloc[i, 0], X_test.iloc[i, 0]], [y_test.iloc[i], y_test_pred[i]], 'k--', linewidth=0.8)  # L√≠nea de error
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.title('Regresi√≥n lineal: Altura vs Peso (Conjunto de prueba)')
plt.legend()
plt.show()
```

```python
# Evaluaci√≥n del modelo en conjunto de prueba
mae = mean_absolute_error(y_test, y_test_pred)
mse = mean_squared_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print("Resultados en el conjunto de prueba:")
print(f"Error absoluto medio (MAE): {mae}")
print(f"Error cuadr√°tico medio (MSE): {mse}")
print(f"Coeficiente de determinaci√≥n (R¬≤): {r2}")
```

```python
# Normalizaci√≥n de los datos de entrada usando MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelo de regresi√≥n lineal
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Predicciones
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

# Evaluaci√≥n del modelo en conjunto de prueba
mae = mean_absolute_error(y_test, y_test_pred)
mse = mean_squared_error(y_test, y_test_pred)
r2 = r2_score(y_test, y_test_pred)

print("Resultados en el conjunto de prueba:")
print(f"Error absoluto medio (MAE): {mae}")
print(f"Error cuadr√°tico medio (MSE): {mse}")
print(f"Coeficiente de determinaci√≥n (R¬≤): {r2}")

# Visualizaci√≥n de los errores
plt.scatter(X_test, y_test, color='blue', label='Datos de prueba')
plt.plot(X_test, y_test_pred, color='red', label='L√≠nea de regresi√≥n')
for i in range(len(X_test)):
    plt.plot([X_test.iloc[i, 0], X_test.iloc[i, 0]], [y_test.iloc[i], y_test_pred[i]], 'k--', linewidth=0.8)  # L√≠nea de error
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.title('Regresi√≥n lineal: Altura vs Peso (Conjunto de prueba)')
plt.legend()
plt.show()
```
