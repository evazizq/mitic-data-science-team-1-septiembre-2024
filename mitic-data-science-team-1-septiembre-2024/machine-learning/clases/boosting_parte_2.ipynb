## Diferencias entre XGBoost y Gradient Boosting



A continuación, se detallan las diferencias clave entre **XGBoost** y el algoritmo genérico de **Gradient Boosting**. Estos dos métodos son ampliamente utilizados en problemas de machine learning y comparten similitudes en su enfoque, pero difieren en implementación y optimizaciones.

---

1. **Gradient Boosting (GB)**

El **Gradient Boosting** es una técnica de machine learning basada en árboles de decisión que construye modelos aditivos de manera secuencial para minimizar una función de pérdida. A grandes rasgos:

**Características principales:**
- **Estrategia general**: Optimiza una función de pérdida mediante el ajuste de modelos base (árboles de decisión) de forma iterativa.
- **Implementación genérica**: Muchas librerías (como Scikit-learn) implementan Gradient Boosting básico, sin optimizaciones específicas.
- **Flexibilidad**: Permite personalizar las funciones de pérdida y configuraciones del modelo, lo que lo hace útil para una variedad de problemas.

**Ventajas:**
- Intuitivo y ampliamente conocido.
- Compatible con distintos tipos de datos.

**Limitaciones:**
- Puede ser lento en datasets grandes.
- No está optimizado para paralelización.
- Tiende a hacer overfitting si no se controlan adecuadamente los parámetros del modelo

---

2. **XGBoost (Extreme Gradient Boosting)**

**XGBoost** es una implementación optimizada y avanzada del Gradient Boosting. Introduce múltiples mejoras en términos de velocidad y rendimiento.

**aracterísticas principales:**
- **Optimización de regularización**: Incluye términos de regularización \( L1 - Lasso regressión\) y \( L2 - Regresión de cresta\) para prevenir sobreajuste.
- **Paralelización**: Diseñado para utilizar múltiples núcleos durante el entrenamiento, acelerando significativamente el proceso.
- **Gestión de memoria**: Eficiente en el manejo de datos, utiliza una estructura de árbol basada en histogramas para optimizar cálculos.
- **Algoritmo avanzado**: Emplea técnicas como el pre-sorting para encontrar divisiones óptimas en los árboles de decisión.
- **Faltantes de datos**: Maneja automáticamente valores nulos o faltantes, asignándolos a la rama más apropiada durante la construcción del árbol.

**Ventajas:**
- Más rápido y eficiente en comparación con Gradient Boosting básico.
- Soporta entrenamiento distribuido, útil en grandes volúmenes de datos.
- Funciona bien en tareas estructuradas (por ejemplo, Kaggle).

**Limitaciones:**
- Puede ser más complejo de configurar debido a la cantidad de hiperparámetros.

---

3. **Diferencias clave entre XGBoost y Gradient Boosting**

| Aspecto                    | Gradient Boosting                    | XGBoost                                   |
|----------------------------|---------------------------------------|-------------------------------------------|
| **Implementación**         | Algoritmo genérico                   | Implementación optimizada de Gradient Boosting |
| **Regularización**         | No tiene regularización explícita    | Regularización \(L1\) y \(L2\) disponibles |
| **Paralelización**         | No paraleliza el entrenamiento       | Altamente paralelizable                   |
| **Eficiencia**             | Más lento en datasets grandes        | Más rápido y eficiente                    |
| **Manejo de valores nulos**| No lo hace automáticamente           | Los gestiona automáticamente              |
| **Popularidad**            | Menos popular en competencia         | Muy popular en competencias como Kaggle   |

---

**Conclusión**

- **Usar Gradient Boosting**: Es adecuado para problemas pequeños o si necesitas una implementación básica y fácil de entender.
- **Usar XGBoost**: Recomendado para problemas más grandes, donde la eficiencia, velocidad y regularización sean críticas.

Por lo tanto, XGBoost puede considerarse como una versión mejorada del Gradient Boosting clásico, diseñado para abordar las limitaciones de este último.


## Diferencias entre LightGBM y Gradient Boosting

A continuación, se describen las diferencias clave entre **LightGBM** y el algoritmo genérico de **Gradient Boosting**. Ambos algoritmos son métodos potentes de machine learning basados en árboles de decisión, pero LightGBM introduce optimizaciones significativas que lo hacen destacar en tareas específicas.

---

1. **Gradient Boosting (GB)**

El **Gradient Boosting** es un algoritmo que construye modelos aditivos de manera secuencial optimizando una función de pérdida. Cada modelo se entrena para corregir los errores residuales del modelo anterior.

**Características principales:**
- **Estrategia general**: Aplica una combinación de árboles de decisión para minimizar la función de pérdida.
- **Implementación estándar**: Librerías como Scikit-learn ofrecen una implementación genérica del algoritmo.
- **Función de pérdida personalizada**: Permite flexibilidad para resolver problemas específicos.

**Ventajas:**
- Intuitivo y ampliamente conocido.
- Es adecuado para conjuntos de datos pequeños o medianos.

**Limitaciones:**
- Carece de optimizaciones avanzadas para grandes volúmenes de datos.
- No es paralelizable en la construcción de árboles.
- Más lento comparado con implementaciones modernas.

---

2. **LightGBM (Light Gradient Boosting Machine)**

**LightGBM** es una implementación moderna y altamente optimizada de Gradient Boosting. Está diseñado específicamente para ser rápido y eficiente en conjuntos de datos grandes.

**Características principales:**
- **División basada en histogramas**: Utiliza histogramas para encontrar divisiones óptimas en lugar de recorrer todos los datos.
- **Crecimiento de árboles leaf-wise**: En lugar de expandir los árboles nivel por nivel, LightGBM prioriza la expansión de las hojas más prometedoras, lo que mejora la precisión.
- **Optimización de memoria**: Su manejo de memoria es eficiente, reduciendo el consumo para conjuntos de datos grandes.
- **Soporte para GPU**: Puede aprovechar la GPU para entrenar modelos aún más rápido.
- **Valores faltantes**: Detecta y maneja automáticamente datos faltantes.

**Ventajas:**
- Entrenamiento más rápido que Gradient Boosting estándar.
- Maneja datasets grandes con facilidad.
- Soporte nativo para valores faltantes y GPU.

**Limitaciones:**
- Puede ser más sensible al sobreajuste debido a su estrategia leaf-wise.
- Más complejo de ajustar en comparación con Gradient Boosting básico.

---

3. **Diferencias clave entre LightGBM y Gradient Boosting**

| Aspecto                    | Gradient Boosting                    | LightGBM                                  |
|----------------------------|---------------------------------------|-------------------------------------------|
| **Implementación**         | Algoritmo genérico                   | Implementación optimizada de Gradient Boosting |
| **División del árbol**     | Basada en nivel (nivel por nivel)     | Basada en hojas (leaf-wise)               |
| **Optimización**           | Básica, sin optimizaciones avanzadas | Usa histogramas y compresión de memoria  |
| **Paralelización**         | No paraleliza el entrenamiento       | Altamente paralelizable                   |
| **Eficiencia**             | Más lento en datasets grandes        | Más rápido y eficiente                    |
| **Manejo de valores nulos**| No lo hace automáticamente           | Los gestiona automáticamente              |
| **Soporte GPU**            | No                                    | Sí                                        |

---

**Conclusión**

- **Usar Gradient Boosting**: Ideal para problemas pequeños o donde no se requieren optimizaciones avanzadas.
- **Usar LightGBM**: Recomendado para problemas grandes, donde la velocidad, el manejo eficiente de memoria y la capacidad de procesar datos complejos son prioritarios.

En resumen, LightGBM es una versión optimizada de Gradient Boosting que sobresale en eficiencia y rendimiento, especialmente para grandes volúmenes de datos.

## Implementación en código

```python
# Importar bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, log_loss
import xgboost as xgb
import lightgbm as lgb

# Verificar la versión de LightGBM
print("Versión de LightGBM:", lgb.__version__)

# Crear un conjunto de datos sintético
np.random.seed(42)
X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_informative=10,
    n_redundant=5,
    n_clusters_per_class=2,
    flip_y=0.1,
    class_sep=1.0,
    random_state=42
)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Configuración de los datos para XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# Configuración de parámetros para XGBoost
xgb_params = {
    "objective": "binary:logistic",
    "eval_metric": "logloss",
    "eta": 0.1,  # Learning rate
    "max_depth": 3,
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    "random_state": 42,
}

# Diccionario para almacenar los resultados de evaluación de XGBoost
xgb_evals_result = {}

# Entrenar el modelo de XGBoost con registro de resultados
xgb_eval = [(dtrain, "train"), (dtest, "test")]
xgb_model = xgb.train(
    params=xgb_params,
    dtrain=dtrain,
    num_boost_round=50,
    evals=xgb_eval,
    evals_result=xgb_evals_result,  # Almacenar resultados de evaluación
)

# Configuración para LightGBM
lgb_train = lgb.Dataset(X_train, label=y_train)
lgb_test = lgb.Dataset(X_test, label=y_test, reference=lgb_train)

lgb_params = {
    "objective": "binary",
    "metric": "binary_logloss",
    "learning_rate": 0.1,
    "num_leaves": 31,
    "max_depth": 3,
    "feature_fraction": 0.8,
    "bagging_fraction": 0.8,
    "bagging_freq": 5,
    "random_state": 42,
}

# Diccionario para almacenar resultados manualmente
lgb_evals_result = {"train": [], "test": []}

# Función de callback para registrar métricas
def record_evaluation(env):
    lgb_evals_result["train"].append(env.evaluation_result_list[0][2])  # Entrenamiento
    lgb_evals_result["test"].append(env.evaluation_result_list[1][2])   # Prueba

# Entrenar el modelo de LightGBM con callback
lgb_model = lgb.train(
    params=lgb_params,
    train_set=lgb_train,
    valid_sets=[lgb_train, lgb_test],  # Conjuntos de validación
    valid_names=["train", "test"],     # Nombres de los conjuntos
    num_boost_round=50,               # Número máximo de iteraciones
    callbacks=[record_evaluation],     # Registrar métricas manualmente
)

# Obtener predicciones y evaluar
xgb_preds = xgb_model.predict(dtest)
xgb_accuracy = accuracy_score(y_test, (xgb_preds > 0.5).astype(int))
xgb_logloss = log_loss(y_test, xgb_preds)

lgb_preds = lgb_model.predict(X_test)
lgb_accuracy = accuracy_score(y_test, (lgb_preds > 0.5).astype(int))
lgb_logloss = log_loss(y_test, lgb_preds)

# Mostrar resultados
print("Resultados de XGBoost:")
print(f"Exactitud: {xgb_accuracy:.4f}")
print(f"Log Loss: {xgb_logloss:.4f}")

print("\nResultados de LightGBM:")
print(f"Exactitud: {lgb_accuracy:.4f}")
print(f"Log Loss: {lgb_logloss:.4f}")

# Graficar la evolución del log loss en entrenamiento y prueba
plt.figure(figsize=(10, 6))
plt.plot(
    xgb_evals_result["train"]["logloss"], label="XGBoost Entrenamiento", color="blue"
)
plt.plot(
    xgb_evals_result["test"]["logloss"], label="XGBoost Prueba", color="red"
)
plt.plot(
    range(len(lgb_evals_result["train"])),
    lgb_evals_result["train"],
    label="LightGBM Entrenamiento",
    linestyle="dashed",
    color="blue",
)
plt.plot(
    range(len(lgb_evals_result["test"])),
    lgb_evals_result["test"],
    label="LightGBM Prueba",
    linestyle="dashed",
    color="red",
)
plt.xlabel("Iteración")
plt.ylabel("Log Loss")
plt.title("Evolución del Log Loss")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
)

# Calcular métricas para XGBoost
xgb_y_pred_binary = (xgb_preds > 0.5).astype(int)

xgb_cm = confusion_matrix(y_test, xgb_y_pred_binary)
print("Matriz de Confusión - XGBoost:")
print(xgb_cm)

xgb_accuracy = accuracy_score(y_test, xgb_y_pred_binary)
xgb_precision = precision_score(y_test, xgb_y_pred_binary)
xgb_recall = recall_score(y_test, xgb_y_pred_binary)
xgb_f1 = f1_score(y_test, xgb_y_pred_binary)
xgb_roc_auc = roc_auc_score(y_test, xgb_preds)

print("\nMétricas de Clasificación - XGBoost:")
print(f"Precisión (Accuracy): {xgb_accuracy:.4f}")
print(f"Precisión (Precision): {xgb_precision:.4f}")
print(f"Recall (Sensibilidad): {xgb_recall:.4f}")
print(f"F1-Score: {xgb_f1:.4f}")
print(f"AUC-ROC: {xgb_roc_auc:.4f}")

# Reporte de clasificación para XGBoost
print("\nReporte de Clasificación - XGBoost:")
print(classification_report(y_test, xgb_y_pred_binary))

# Calcular métricas para LightGBM
lgb_y_pred_binary = (lgb_preds > 0.5).astype(int)

lgb_cm = confusion_matrix(y_test, lgb_y_pred_binary)
print("\nMatriz de Confusión - LightGBM:")
print(lgb_cm)

lgb_accuracy = accuracy_score(y_test, lgb_y_pred_binary)
lgb_precision = precision_score(y_test, lgb_y_pred_binary)
lgb_recall = recall_score(y_test, lgb_y_pred_binary)
lgb_f1 = f1_score(y_test, lgb_y_pred_binary)
lgb_roc_auc = roc_auc_score(y_test, lgb_preds)

print("\nMétricas de Clasificación - LightGBM:")
print(f"Precisión (Accuracy): {lgb_accuracy:.4f}")
print(f"Precisión (Precision): {lgb_precision:.4f}")
print(f"Recall (Sensibilidad): {lgb_recall:.4f}")
print(f"F1-Score: {lgb_f1:.4f}")
print(f"AUC-ROC: {lgb_roc_auc:.4f}")

# Reporte de clasificación para LightGBM
print("\nReporte de Clasificación - LightGBM:")
print(classification_report(y_test, lgb_y_pred_binary))
```
