## Sesgo y Varianza en el Aprendizaje Profundo

### ¿Qué es el Sesgo?

El sesgo se refiere a los supuestos que un modelo hace sobre los datos. En términos prácticos, es la capacidad del modelo para captar patrones complejos en el conjunto de datos.

- **¿Cómo se usa?**  
  Un modelo con alto sesgo simplifica demasiado la relación entre variables, lo que puede ser útil en escenarios donde los datos son simples o muy ruidosos. Sin embargo, tiende a subajustarse (no aprende suficientemente bien de los datos de entrenamiento).

- **Ejemplo:**  
  Un modelo de regresión lineal intenta predecir la relación entre dos variables, pero si la relación es no lineal, el modelo no podrá ajustarse correctamente.  
  $$ \hat{y} = w_0 + w_1x $$
  Aquí, la relación lineal impone un sesgo alto.

---

### ¿Qué es la Varianza?

La varianza mide cuánto cambia un modelo al entrenarse con diferentes subconjuntos de datos. Un modelo con alta varianza es muy sensible a los datos de entrenamiento, lo que puede llevar a sobreajuste.

- **¿Cómo se usa?**  
  Un modelo con alta varianza puede ajustarse extremadamente bien a los datos de entrenamiento, incluyendo ruidos y peculiaridades. Esto resulta en un mal desempeño en datos no vistos.

- **Ejemplo:**  
  Una red neuronal con muchas capas y sin regularización aprende patrones complejos, pero también puede memorizar datos ruidosos.  
  $$ \hat{y} = f(x; \theta) $$
  Donde $f$ es una red neuronal con parámetros $\theta$.

---

### El Dilema del Sesgo-Varianza

Este dilema implica encontrar un balance entre bajo sesgo (capturar patrones complejos) y baja varianza (generalizar bien a nuevos datos). 

- **Modelo de bajo sesgo y alta varianza:**  
  Captura demasiados detalles de los datos de entrenamiento, incluyendo ruido.  
  - **Consecuencia:** Sobreajuste.
  
- **Modelo de alto sesgo y baja varianza:**  
  Simplifica en exceso los datos.  
  - **Consecuencia:** Subajuste.
  
- **Modelo equilibrado:**  
  Generaliza bien a datos nuevos.

---

### Técnicas para Manejar el Sesgo y la Varianza

#### **Regularización L1 y L2**

##### **Regularización L1**

---

- **¿Qué es?**  
  La regularización L1 (también conocida como Lasso) es una técnica que penaliza la suma de los valores absolutos de los coeficientes del modelo. Su principal característica es que puede forzar algunos coeficientes a ser exactamente cero, lo que resulta en un modelo más simple y esparso.

- **¿Cómo se usa?**  
  Se agrega un término de penalización proporcional a la norma $L_1$ de los coeficientes a la función de pérdida del modelo.

- **Ejemplo:**  
  La función de pérdida con regularización L1 es:  
  $$ \mathcal{L}(\theta) = \frac{1}{N} \sum_{i=1}^N \text{loss}(y_i, \hat{y}_i) + \lambda \|\theta\|_1 $$  
  Donde $\|\theta\|_1 = \sum_{j} |\theta_j|$ y $\lambda$ controla la fuerza de la penalización.  

  Esta técnica es útil cuando se busca seleccionar automáticamente características relevantes eliminando las irrelevantes (coeficientes cero).

---

##### **Regularización L2**

---
- **¿Qué es?**  
  La regularización L2 (también conocida como Ridge) penaliza la suma de los cuadrados de los coeficientes del modelo. No fuerza coeficientes a cero, sino que reduce sus valores, lo que ayuda a prevenir sobreajuste mientras conserva toda la información.

- **¿Cómo se usa?**  
  Se agrega un término de penalización proporcional a la norma $L_2$ de los coeficientes a la función de pérdida del modelo.

- **Ejemplo:**  
  La función de pérdida con regularización L2 es:  
  $$ \mathcal{L}(\theta) = \frac{1}{N} \sum_{i=1}^N \text{loss}(y_i, \hat{y}_i) + \lambda \|\theta\|_2^2 $$  
  Donde $\|\theta\|_2^2 = \sum_{j} \theta_j^2$ y $\lambda$ controla la fuerza de la penalización.  

  Esta técnica es útil para modelos con muchos parámetros y alta complejidad, ayudando a estabilizar el aprendizaje y a reducir la varianza.
---

##### **Diferencias clave entre L1 y L2**

---

| Aspecto             | Regularización L1              | Regularización L2              |
|---------------------|--------------------------------|--------------------------------|
| Penalización        | Suma de valores absolutos ($\|\theta\|_1$) | Suma de cuadrados ($\|\theta\|_2^2$) |
| Efecto en coeficientes | Fuerza algunos coeficientes a cero (esparcidad) | Reduce el valor de todos los coeficientes |
| Uso principal       | Selección de características   | Manejar modelos complejos o multicolinealidad |
| Interpretabilidad   | Mayor (por la esparcidad)      | Menor (mantiene todos los coeficientes) |

Ambas técnicas son útiles dependiendo del problema y los datos, y se pueden combinar en una penalización **ElasticNet** que mezcla L1 y L2.

---

#### Dropout

- **¿Qué es?**  
  Técnica para redes neuronales que desactiva neuronas aleatoriamente durante el entrenamiento.

- **¿Cómo se usa?**  
  Se define un porcentaje de neuronas a "apagar" en cada capa.

- **Ejemplo:**  
  En una red con 100 neuronas, un dropout del 50% apaga 50 neuronas en cada iteración.

#### Early Stopping

---
- **¿Qué es?**  
  Detiene el entrenamiento cuando el rendimiento en un conjunto de validación deja de mejorar.

- **¿Cómo se usa?**  
  Se evalúa el modelo después de cada época. Si no hay mejoras durante varias épocas consecutivas, el entrenamiento se detiene.

- **Ejemplo:**  
  Si el error de validación deja de disminuir después de 10 épocas, se detiene el entrenamiento.
---

#### Validación Cruzada

---
- **¿Qué es?**  
  Técnica que divide los datos en múltiples subconjuntos para entrenar y validar el modelo repetidamente.

- **¿Cómo se usa?**  
  Dividiendo los datos en $k$ pliegues, entrenando en $k-1$ pliegues y validando en el pliegue restante.

- **Ejemplo:**  
  Con validación cruzada de 5 pliegues, los datos se dividen en 5 partes, y cada parte se usa una vez como conjunto de validación.

### Ejemplo con código

#### Descargamos datos

```python
import os
from kaggle.api.kaggle_api_extended import KaggleApi

# Inicializar API
api = KaggleApi()
api.authenticate()

# Directorio de destino
destination_dir = "../data/manjitbaishya001"
os.makedirs(destination_dir, exist_ok=True)

# Descargar el dataset
api.dataset_download_files("manjitbaishya001/sonar-data", path=destination_dir, unzip=True)

print(f"Dataset descargado y descomprimido en: {destination_dir}")
```

#### Creación y entrenamiento de modelos con diferentes técnicas

```python
from sklearn.metrics import roc_auc_score
from tensorflow.keras.layers import Dropout
# Importar librerías necesarias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2,l1
from tensorflow.keras.callbacks import EarlyStopping

# Función para crear y entrenar modelos con diferentes configuraciones
def train_model(config, X_train, y_train, X_val, y_val):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu', kernel_regularizer=config.get('regularizer')))
    
    if config.get('dropout'):
        model.add(Dropout(config['dropout']))
        
    model.add(Dense(32, activation='relu', kernel_regularizer=config.get('regularizer')))
    model.add(Dense(2, activation='softmax'))

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    callbacks = []
    if config.get('early_stopping'):
        callbacks.append(EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True))

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=500,
        batch_size=16,
        callbacks=callbacks,
        verbose=0
    )
    return model, history

# Configuraciones para los experimentos
configs = {
    'baseline': {},
    'dropout': {'dropout': 0.2},
    'early_stopping': {'early_stopping': True},
    'l2_regularization': {'regularizer': l2(0.01)},
    'l1_regularization': {'regularizer': l1(0.01)}
}
```

```python
# Cargar el dataset desde la carpeta '../data/manjitbaishya001'
data_path = "../data/manjitbaishya001/sonar_data.csv"
data = pd.read_csv(data_path)

# Separar características (X) y etiquetas (y)
X = data.iloc[:, :-1].values
y = pd.get_dummies(data.iloc[:, -1]).values  # Convertir etiquetas a formato one-hot encoding

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Dividir datos de entrenamiento en entrenamiento y validación
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Entrenar modelos con diferentes configuraciones
models = {}
histories = {}
for name, config in configs.items():
    print(f"Entrenando modelo: {name}")
    model, history = train_model(config, X_train, y_train, X_val, y_val)
    models[name] = model
    histories[name] = history
```

#### Evaluación de los modelos

```python
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_fscore_support

# Evaluación en el conjunto de prueba
evaluation_results = {}

for name, model in models.items():
    y_pred = np.argmax(model.predict(X_test), axis=1)
    y_true = np.argmax(y_test, axis=1)

    # Métricas de evaluación
    accuracy = accuracy_score(y_true, y_pred)
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')
    roc_auc = roc_auc_score(y_test, model.predict(X_test), multi_class='ovr')

    evaluation_results[name] = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'roc_auc': roc_auc
    }

# Imprimir los resultados
for name, metrics in evaluation_results.items():
    print(f"Resultados para el modelo {name}:")
    for metric, value in metrics.items():
        print(f"  {metric}: {value:.4f}")
```

#### Matriz de confusión

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

for name, model in models.items():
    y_pred = np.argmax(model.predict(X_test), axis=1)
    y_true = np.argmax(y_test, axis=1)
    
    ConfusionMatrixDisplay.from_predictions(y_true, y_pred, cmap='Blues', display_labels=["Clase 0", "Clase 1"])
    plt.title(f"Matriz de Confusión - {name}")
    plt.show()
```

#### Visualización de la pérdida y curvas ROC-AUC

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
# Graficar la diferencia en la pérdida entre entrenamiento y validación por época
plt.figure(figsize=(12, 6))
for name, history in histories.items():
    training_loss = history.history['loss']
    validation_loss = history.history['val_loss']
    difference =  np.array(validation_loss) - np.array(training_loss)
    plt.plot(difference, label=f"{name} - Diferencia (Validación - Entrenamiento)")

plt.axhline(0, color='k', linestyle='--', linewidth=1, label="Diferencia = 0")
plt.title("Diferencia en la pérdida (Validación - Entrenamiento) por época")
plt.xlabel("Épocas")
plt.ylabel("Diferencia en la pérdida")
plt.legend()
plt.show()

# Graficar curvas ROC-AUC
plt.figure(figsize=(12, 6))
for name, model in models.items():
    y_pred_prob = model.predict(X_test)[:, 1]  # Probabilidad para la clase positiva
    fpr, tpr, _ = roc_curve(y_test[:, 1], y_pred_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.4f})")

plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")
plt.title("Curvas ROC-AUC")
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.legend()
plt.show()
```

#### Precisión y pérdida en entrenamiento y validación

```python
plt.figure(figsize=(12, 6))
for name, history in histories.items():
    plt.plot(history.history['accuracy'], label=f"{name} - Precisión (Entrenamiento)")
    plt.plot(history.history['val_accuracy'], linestyle='--', label=f"{name} - Precisión (Validación)")

plt.title("Precisión durante el entrenamiento y validación")
plt.xlabel("Épocas")
plt.ylabel("Precisión")
plt.legend()
plt.show()
```

#### Histograma de probabilidades predichas

```python
import seaborn as sns

# Crear un gráfico de distribución para las probabilidades predichas de cada modelo
plt.figure(figsize=(12, 6))
for name, model in models.items():
    y_pred_prob = model.predict(X_test)[:, 1]  # Probabilidades de la clase positiva
    sns.kdeplot(y_pred_prob, label=f"{name}", fill=True, alpha=0.5)

plt.title("Distribución de probabilidades predichas (Clase positiva)")
plt.xlabel("Probabilidad")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```

```python
import seaborn as sns

# Seleccionar los modelos específicos
models_to_plot = ['baseline','dropout']

# Crear un gráfico de distribución para las probabilidades predichas de los modelos seleccionados
plt.figure(figsize=(12, 6))
for name in models_to_plot:
    y_pred_prob = models[name].predict(X_test)[:, 1]  # Probabilidades de la clase positiva
    sns.kdeplot(y_pred_prob, label=f"{name}", fill=True, alpha=0.5)

plt.title("Distribución de probabilidades predichas (Clase positiva) - Modelos seleccionados")
plt.xlabel("Probabilidad")
plt.ylabel("Densidad")
plt.legend()
plt.show()
```

#### Curva de precisión-recall

```python
from sklearn.metrics import precision_recall_curve

plt.figure(figsize=(12, 6))
for name, model in models.items():
    y_pred_prob = model.predict(X_test)[:, 1]
    precision, recall, _ = precision_recall_curve(y_test[:, 1], y_pred_prob)
    plt.plot(recall, precision, label=f"{name}")

plt.title("Curva de Precisión-Recall")
plt.xlabel("Recall")
plt.ylabel("Precisión")
plt.legend()
plt.show()
```
