## Fundamentos de la primera generación de algoritmos que usan Boosting

### **Conceptos**

#### ***Learners*** débiles vs ***learners*** fuertes

Un **learner débil** (o modelo débil) es un algoritmo de aprendizaje cuya precisión en la predicción es apenas mejor que el azar (en clasificación, ligeramente superior al 50% para un problema binario). Aunque un learner débil por sí solo no es muy útil, puede combinarse con otros learners débiles mediante técnicas como el **boosting** para crear un modelo fuerte.

Por el contrario, un **learner fuerte** (o modelo fuerte) es un modelo que tiene un rendimiento significativamente mejor que el azar y puede tomar decisiones complejas por sí mismo.

---

**1. ¿Qué es un Learner Débil?**

Un learner débil cumple las siguientes características:
- Tiene un desempeño marginalmente mejor que el azar.
- Es simple y rápido de entrenar.
- Puede fallar en capturar patrones complejos de los datos.

Ejemplo clásico:
- Un **árbol de decisión de baja profundidad** (por ejemplo, un árbol con una profundidad máxima de 1, también llamado "stump").

---

**2. Ejemplos de Learners Débiles**

| **Learner Débil**                       | **Descripción**                                                                 |
|-----------------------------------------|---------------------------------------------------------------------------------|
| Árbol de Decisión de Baja Profundidad   | Un árbol que realiza una sola división o unas pocas divisiones.                |
| Regresión Lineal                        | Un modelo que ajusta una línea a los datos (solo útil en problemas lineales).  |
| Regresión Logística Simple              | Un modelo que predice probabilidades lineales para clasificación binaria.      |
| Clasificación Basada en Reglas Simples  | Ejemplo: Dividir en clases usando un umbral para una característica.           |
| Naive Bayes                             | Un modelo probabilístico basado en independencia de características.           |

---

**3. ¿Qué es un Learner Fuerte?**

Un learner fuerte es un modelo que puede tomar decisiones complejas de manera efectiva. Estos modelos tienen la capacidad de aprender patrones complejos de los datos y generalizar bien a datos no vistos.

Ejemplo clásico:
- **Redes Neuronales Profundas** o un **Random Forest** bien entrenado.

---

**4. Comparación entre Learners Débiles y Fuertes**

| **Característica**         | **Learner Débil**                                      | **Learner Fuerte**                           |
|-----------------------------|-------------------------------------------------------|----------------------------------------------|
| **Capacidad de Predicción** | Ligeramente mejor que el azar                         | Alta precisión y generalización.             |
| **Complejidad**             | Muy simple                                            | Complejo y capaz de capturar relaciones no lineales. |
| **Rendimiento Individual**  | Limitado, necesita ser combinado con otros models     | Suficiente para resolver problemas por sí mismo. |
| **Tiempo de Entrenamiento** | Muy rápido                                            | Más lento debido a su complejidad.           |
| **Ejemplos Comunes**        | Árboles de decisión poco profundos, Naive Bayes       | Redes Neuronales, Random Forest, XGBoost.    |

---

**5. Relación Entre Learners Débiles y Técnicas de Ensemble**

Los learners débiles son la base de técnicas como:
1. **Boosting**:
   - Combina múltiples learners débiles (por ejemplo, árboles de decisión de baja profundidad) en un modelo fuerte, corrigiendo errores iterativamente. Ejemplo: Gradient Boosting.
2. **Bagging**:
   - Usa múltiples learners débiles de manera independiente y promedia sus predicciones. Ejemplo: Random Forest.

Estas técnicas transforman modelos débiles en modelos fuertes, aumentando su capacidad de predicción.

---

**6. ¿Por Qué Son Importantes los Learners Débiles?**

Aunque los learners débiles tienen un rendimiento limitado, son valiosos porque:
- Son rápidos de entrenar.
- Pueden combinarse en técnicas de ensemble para resolver problemas complejos.
- Su simplicidad reduce el riesgo de sobreajuste en cada modelo individual.

#### **El Gradiente Descendente: Un Método de Optimización**

El **gradiente descendente** es un algoritmo iterativo ampliamente utilizado para optimizar funciones en problemas de aprendizaje automático, especialmente en redes neuronales y modelos de regresión. Su objetivo es encontrar el mínimo de una función de pérdida ajustando los parámetros del modelo.

---

**1. ¿Qué es el Gradiente Descendente?**

El gradiente descendente es un método de optimización basado en la idea de que los parámetros de un modelo ($\theta$) deben ajustarse en la dirección del gradiente negativo de una función de pérdida ($L(\theta)$) para minimizarla.

La función de pérdida mide el error entre las predicciones del modelo y los valores reales. Minimizar esta función ayuda al modelo a mejorar su rendimiento.

---

**2. ¿Cómo Funciona el Gradiente Descendente?**

El algoritmo sigue estos pasos:

1. **Inicialización de Parámetros**:
   - Se inicializan los parámetros ($\theta$) con valores aleatorios o un punto de partida predeterminado.

2. **Cálculo del Gradiente**:
   - Se calcula el gradiente de la función de pérdida con respecto a los parámetros ($\nabla L(\theta)$).
   - El gradiente indica la dirección de mayor aumento de la función. El gradiente negativo apunta hacia el mínimo.

3. **Actualización de los Parámetros**:
   - Los parámetros se actualizan moviéndose en la dirección opuesta al gradiente, escalados por un factor llamado **tasa de aprendizaje** ($\eta$):
     $$
     \theta = \theta - \eta \cdot \nabla L(\theta)
     $$
     - $\eta$: Controla el tamaño del paso. Valores altos pueden llevar a una convergencia inestable, mientras que valores muy bajos ralentizan el proceso.

4. **Repetición**:
   - El proceso se repite hasta que se cumpla un criterio de parada, como un número máximo de iteraciones o cuando los cambios en la función de pérdida sean insignificantes.

---

**3. Tipos de Gradiente Descendente**

1. **Batch Gradient Descent**:
   - Usa el conjunto completo de datos para calcular el gradiente.
   - Más preciso, pero computacionalmente costoso para datasets grandes.

2. **Stochastic Gradient Descent (SGD)**:
   - Actualiza los parámetros usando un único ejemplo a la vez.
   - Más rápido, pero puede ser ruidoso y no converger fácilmente.

3. **Mini-Batch Gradient Descent**:
   - Combina lo mejor de ambos. Usa pequeños lotes de datos para calcular el gradiente.
   - Es el método más comúnmente utilizado en la práctica.

---

**4. Ejemplo Intuitivo**

Supongamos que estás en una montaña y quieres llegar al valle más bajo. El gradiente descendente te indica la dirección más pronunciada hacia abajo (gradiente negativo). A medida que avanzas, ajustas tus pasos ($\eta$) para no avanzar demasiado rápido ni demasiado lento, buscando el punto más bajo (mínimo).

---

**5. Ventajas del Gradiente Descendente**

- **Eficiencia**: Es computacionalmente eficiente para funciones diferenciables.
- **Escalabilidad**: Puede manejar datasets grandes (usando mini-batches).
- **Generalización**: Ayuda a encontrar soluciones optimizadas en problemas complejos.

---

**6. Limitaciones del Gradiente Descendente**

- **Elección de la Tasa de Aprendizaje**: Un $\eta$ inadecuado puede llevar a una convergencia inestable o lenta.
- **Mínimos Locales**: Puede quedarse atrapado en mínimos locales para funciones no convexas.
- **Condición Inicial**: Los valores iniciales pueden influir en el resultado.

---

**7. Visualización Matemática**

Para una función de pérdida $L(\theta)$, el gradiente descendente realiza los siguientes pasos:
1. Calcula el gradiente: $\nabla L(\theta)$.
2. Actualiza los parámetros:
   $$
   \theta_{t+1} = \theta_t - \eta \cdot \nabla L(\theta_t)
   $$


##### Implementación en código

```python
import numpy as np

# Función de pérdida: f(x) = x^2
def loss_function(x):
    return x**2

# Derivada de la función: f'(x) = 2x
def gradient(x):
    return 2 * x

# Gradiente Descendente
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    history = [x]
    for _ in range(iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        history.append(x)
    return x, history

# Parámetros
start = 10  # Punto inicial
lr = 0.1    # Tasa de aprendizaje
iters = 50  # Número de iteraciones

# Ejecutar Gradiente Descendente
final_x, path = gradient_descent(start, lr, iters)

# Resultados
print(f"Punto mínimo encontrado: {final_x}")

# Graficar el proceso
import matplotlib.pyplot as plt

x_vals = np.linspace(-10, 10, 100)
y_vals = loss_function(x_vals)

plt.plot(x_vals, y_vals, label="Función de pérdida (x^2)")
plt.scatter(path, [loss_function(x) for x in path], color="red", label="Iteraciones")
plt.title("Gradiente Descendente")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

#### **El Método Aditivo en los Algoritmos de Boosting**

El **método aditivo** es el enfoque central de los algoritmos de boosting, como AdaBoost y Gradient Boosting. Su principal idea es construir un modelo fuerte combinando iterativamente múltiples **weak learners** (modelos débiles), donde cada nuevo modelo se suma al modelo anterior para corregir los errores acumulados.

---

**1. ¿Qué es el Método Aditivo?**

El método aditivo utiliza una combinación secuencial de modelos para minimizar una función de pérdida. En cada iteración:
1. Se ajusta un nuevo modelo ($h_t(x)$) para corregir los errores del modelo actual ($F_{t-1}(x)$).
2. Este modelo se suma al modelo existente, escalado por un peso o tasa de aprendizaje ($\eta$).

La idea clave es que cada modelo adicional compensa las deficiencias del conjunto actual, mejorando gradualmente el rendimiento.

---

2. Representación Matemática**

Un modelo aditivo puede representarse como:

$$
F_t(x) = F_{t-1}(x) + \eta \cdot h_t(x)
$$

Donde:
- $F_t(x)$: El modelo combinado hasta la iteración $t$.
- $F_{t-1}(x)$: El modelo combinado hasta la iteración anterior ($t-1$).
- $h_t(x)$: El weak learner ajustado en la iteración actual.
- $\eta$: Tasa de aprendizaje (*learning rate*), que controla la contribución de $h_t(x)$.

---

**3. Pasos del Método Aditivo**

**Paso 1: Inicialización**
Se empieza con un modelo inicial $F_0(x)$, que suele ser una predicción constante:
- Para regresión: La media de los valores objetivo ($y_i$).
- Para clasificación: La probabilidad logarítmica ($\log(p/(1-p))$).

$$
F_0(x) = \text{predicción inicial}
$$

---

**Paso 2: Cálculo de Residuos o Errores**
En cada iteración, se calculan los residuos $r_i$, que representan los errores que el nuevo weak learner intentará corregir:

$$
r_i = y_i - F_{t-1}(x_i)
$$

---

**Paso 3: Ajuste del Weak Learner**
Se entrena un **weak learner** ($h_t(x)$) para predecir los residuos calculados en el paso anterior.

---

**Paso 4: Actualización del Modelo**
El modelo se actualiza sumando una versión escalada del weak learner:

$$
F_t(x) = F_{t-1}(x) + \eta \cdot h_t(x)
$$

---

**Paso 5: Repetición**
Este proceso se repite hasta alcanzar un número deseado de iteraciones o hasta que la función de pérdida deje de mejorar significativamente.

---

**4. Métodos de Boosting que Usan el Método Aditivo**

**AdaBoost**
Combina weak learners ajustando pesos en las muestras:

$$
F_t(x) = F_{t-1}(x) + \alpha_t \cdot h_t(x)
$$

Aquí, $\alpha_t$ es un peso basado en el error del weak learner.

---

**Gradient Boosting**
Usa el descenso por gradiente para minimizar directamente una función de pérdida ($L$):

$$
F_t(x) = F_{t-1}(x) - \eta \cdot \nabla L(F_{t-1}(x))
$$

---

**5. Ventajas del Método Aditivo**

- **Flexibilidad**:
  Puede ajustarse a diferentes funciones de pérdida para problemas de regresión, clasificación y ranking.

- **Precisión**:
  Al corregir iterativamente los errores, alcanza altos niveles de precisión.

- **Adaptabilidad**:
  Se adapta a problemas complejos aprendiendo patrones gradualmente.

---

**6. Limitaciones del Método Aditivo**

- **Lento en Entrenamiento**:
  El enfoque secuencial puede ser costoso para datasets grandes.

- **Riesgo de Sobreajuste**:
  Sin una buena regularización (como un learning rate bajo), puede sobreajustarse a los datos de entrenamiento.

- **Sensibilidad a los Weak Learners**:
  Depende de la calidad de los weak learners; modelos débiles o mal configurados pueden limitar el rendimiento.

---

**Resumen Visual**

| **Paso**                    | **Descripción**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| Inicialización              | Comienza con un modelo base $F_0(x)$.                                         |
| Cálculo de Residuos         | Los residuos representan los errores actuales ($r_i = y_i - F_{t-1}(x_i)$).   |
| Weak Learner                | Se entrena un modelo débil para predecir los residuos ($h_t(x)$).             |
| Actualización del Modelo    | Se actualiza el modelo sumando una versión ponderada del weak learner ($\eta \cdot h_t(x)$). |
| Repetición                  | Se repiten los pasos hasta alcanzar el número de iteraciones deseado.          |



#### **Mecánica de los algoritmos de boosting**

El **boosting** es una técnica de ensemble que combina múltiples **learners débiles** (modelos simples con un rendimiento marginal) para construir un **learner fuerte** (modelo altamente preciso). Su enfoque principal es entrenar los learners secuencialmente, donde cada uno intenta corregir los errores del anterior, lo que resulta en un modelo más preciso.

---

**1. Paso a Paso de los Algoritmos de Boosting**

El proceso general del boosting puede describirse en los siguientes pasos:

1. **Inicialización del Modelo Base**:
   - Comienza con un modelo inicial ($F_0(x)$), que puede ser una predicción simple, como la media de los valores en regresión o la probabilidad logarítmica en clasificación.

   Ejemplo:
   - Regresión: $$(F_0(x) = \bar{y})$$
   - Clasificación: $$F_0(x) = \log\left(\frac{p}{1-p}\right)$$

---

2. **Cálculo del Error o Residuos**:
   - Para cada iteración \(t\), se calcula el error o los residuos ($r_i$) del modelo actual ($F_{t-1}(x)$):
     $$
     r_i = y_i - F_{t-1}(x_i)
     $$
   - En boosting de clasificación, puede implicar ajustar los pesos de las muestras mal clasificadas.

---

3. **Entrenamiento del Weak Learner**:
   - Un **learner débil** ($h_t(x)$) es entrenado para ajustar los residuos calculados en el paso anterior.
   - El objetivo del weak learner es centrarse en los errores actuales del modelo.

---

4. **Cálculo de la Contribución del Weak Learner**:
   - Se calcula un factor de escala ($\alpha_t$) para ajustar la contribución del weak learner al modelo final.
     - En **AdaBoost**, esto está relacionado con el error ponderado ($\epsilon_t$) del weak learner:
       $$
       \alpha_t = \ln\left(\frac{1 - \epsilon_t}{\epsilon_t}\right)
       $$
     - En **Gradient Boosting**, $\alpha_t$ suele ser la tasa de aprendizaje ($\eta$).

---

5. **Actualización del Modelo**:
   - El modelo actual ($F_{t-1}(x)$) se actualiza sumando el weak learner escalado por su contribución ($\alpha_t$):
     $$
     F_t(x) = F_{t-1}(x) + \alpha_t \cdot h_t(x)
     $$
   - En el caso de Gradient Boosting, el modelo intenta minimizar directamente una función de pérdida ($L$) ajustando el gradiente:
     $$
     F_t(x) = F_{t-1}(x) - \eta \cdot \nabla L(F_{t-1}(x))
     $$

---

6. **Repetición del Proceso**:
   - Los pasos de cálculo de residuos, entrenamiento del weak learner y actualización del modelo se repiten durante un número definido de iteraciones o hasta que el error sea suficientemente pequeño.

---

**2. Mecanismo General del Boosting**

| **Paso**               | **Descripción**                                                                 |
|------------------------|---------------------------------------------------------------------------------|
| Inicialización         | Comienza con un modelo base $F_0(x)$.                                         |
| Cálculo de Residuos    | Determina los errores del modelo actual ($r_i = y_i - F_{t-1}(x_i)$).         |
| Weak Learner           | Entrena un modelo simple para ajustar los residuos ($h_t(x)$).               |
| Actualización del Modelo | Ajusta el modelo sumando el weak learner escalado ($F_t(x) = F_{t-1}(x) + \alpha_t \cdot h_t(x)$). |
| Repetición             | Repite el proceso hasta alcanzar un criterio de parada.                        |

---

**3. Tipos de Algoritmos de Boosting**

1. **AdaBoost**:
   - Pondera las muestras mal clasificadas, dándoles mayor importancia en el entrenamiento de los siguientes weak learners.
   - Combina weak learners usando un peso calculado en función del error:
     $$
     F_t(x) = F_{t-1}(x) + \alpha_t \cdot h_t(x)
     $$

2. **Gradient Boosting**:
   - Minimiza directamente una función de pérdida ajustando el gradiente:
     $$
     F_t(x) = F_{t-1}(x) - \eta \cdot \nabla L(F_{t-1}(x))
     $$
   - Es más flexible, ya que permite usar diferentes funciones de pérdida (p. ej., MSE, log loss).

3. **XGBoost y LightGBM**:
   - Son optimizaciones de Gradient Boosting que incluyen regularización, manejo eficiente de memoria y paralelismo.

---

**4. Ventajas del Boosting**

- **Precisión Alta**:
  - Produce modelos fuertes que generalmente superan a otros métodos de ensemble como el bagging.
- **Flexibilidad**:
  - Admite funciones de pérdida personalizadas.
- **Adaptación al Error**:
  - Los weak learners se enfocan en corregir los errores acumulados.

---

**5. Limitaciones del Boosting**

- **Riesgo de Sobreajuste**:
  - Puede sobreajustarse fácilmente si el modelo no se regulariza adecuadamente.
- **Sensibilidad al Ruido**:
  - Datos ruidosos pueden desviar el entrenamiento de los weak learners.
- **Lento en Entrenamiento**:
  - El entrenamiento secuencial puede ser computacionalmente costoso.

---

**6. Visualización del Proceso**

1. Inicialización: $$F_0(x)$$ se establece como una predicción inicial (p. ej., la media en regresión).
2. Weak Learner: Cada weak learner ajusta los errores de $$F_{t-1}(x)$$.
3. Combinación: Los weak learners se combinan iterativamente para construir el modelo final $$F_T(x)$$.

Este mecanismo es la base de algoritmos modernos como **XGBoost** y **LightGBM**.

### **Algoritmo Adaboost**

##### **Características**

* Se inicia con un muestreo de datos (es decir, una sub-muestra) que es recogida mediante muestreo aleatorio con reemplazo

* Va ajustando los pesos de los learners débiles a medida que el algoritmo va entrenando y 

En AdaBoost, los pesos de los weak learners ($\alpha_t$) se calculan con la fórmula:

$$
\alpha_t = \ln\left(\frac{1 - \epsilon_t}{\epsilon_t}\right)
$$

Donde:

- $\epsilon_t$ es el error de clasificación ponderado del weak learner.

**SAMME vs SAMME.R en AdaBoost**

AdaBoost admite dos variantes principales: **SAMME** (Discrete) y **SAMME.R** (Real). Ambos son algoritmos de boosting, pero difieren en cómo calculan y combinan las contribuciones de los weak learners.

---

**1. SAMME (Stagewise Additive Modeling using a Multi-class Exponential loss)**
- **Tipo**: Boosting discreto.
- **Salidas de los weak learners**:
  - Utiliza **predicciones categóricas** (clases) de los weak learners.
  - Los weak learners devuelven directamente la clase asignada (por ejemplo, `0` o `1`).
- **Cálculo de los pesos de los weak learners**:
  - Basado en el **error de clasificación ponderado** (\(\epsilon_t\)) de cada weak learner:
    \[
    \alpha_t = \ln\left(\frac{1 - \epsilon_t}{\epsilon_t}\right) + \ln(K - 1)
    \]
    Donde \(K\) es el número de clases.
- **Predicción final**:
  - La predicción final se obtiene mediante una votación ponderada por los pesos (\(\alpha_t\)) de los weak learners.

---

**2. SAMME.R (Real)**
- **Tipo**: Boosting real.
- **Salidas de los weak learners**:
  - Utiliza las **probabilidades de clase** estimadas por los weak learners en lugar de predicciones categóricas.
  - Requiere que los weak learners devuelvan probabilidades (por ejemplo, árboles con `predict_proba`).
- **Cálculo de los pesos de los weak learners**:
  - Los pesos se derivan de las probabilidades en lugar del error de clasificación:
    \[
    F_t(x) = \sum_{t=1}^T \alpha_t h_t(x)
    \]
    Donde \(F_t(x)\) es el puntaje acumulado para cada clase y \(h_t(x)\) las probabilidades del weak learner.
- **Predicción final**:
  - Utiliza la suma ponderada de las probabilidades de los weak learners para determinar la clase final.

---

**Diferencias Clave**
| Característica         | **SAMME**                  | **SAMME.R**                 |
|------------------------|----------------------------|-----------------------------|
| **Tipo**              | Boosting discreto          | Boosting real               |
| **Salidas**           | Predicciones categóricas   | Probabilidades              |
| **Cálculo de pesos**  | Basado en el error (\(\epsilon_t\)) | Basado en probabilidades    |
| **Convergencia**      | Requiere más iteraciones   | Converge más rápido         |
| **Compatibilidad**    | Funciona con cualquier weak learner | Requiere `predict_proba` en el modelo base |

---

**Conclusión**
- Usa **SAMME.R** cuando los weak learners puedan proporcionar probabilidades, ya que converge más rápido y utiliza información adicional.
- Usa **SAMME** cuando los weak learners solo devuelvan predicciones de clase y no probabilidades.

##### **Efectos de la escala de los Errores**

Si los errores ($\epsilon_t$) están en un rango similar (por ejemplo, entre 0.4 y 0.5 en tu tabla), los valores de:

$$
\ln\left(\frac{1 - \epsilon_t}{\epsilon_t}\right)
$$

pueden converger hacia valores similares, resultando en pesos iguales.

**Ejemplo:**
Para errores cercanos a 0.5:

$$
\ln\left(\frac{1 - 0.46}{0.46}\right) \approx 0.08, \quad \ln\left(\frac{1 - 0.48}{0.48}\right) \approx 0.04
$$

La diferencia en los pesos es mínima.

---

**Solución a este problema**
Ajusta los weak learners y/o introduce más variabilidad en los datos para diversificar los errores y/o aumenta el número de iteraciones.

##### **Efecto del algoritmo que usa Adaboost sobre los pesos**

En SAMME.R, los weak learners no tienen pesos explícitos asignados para combinarse en el ensemble final, sino que su salida es utilizada directamente como una probabilidad para ponderar las predicciones finales. Por eso, los valores en estimator_weights_ no varían.

##### **Ajustamos el modelo para mostrar el efecto deseado en cambio de pesos**

Esto se logra ajustando los datos, los weak learners y el número de iteraciones para garantizar que los errores varíen lo suficiente entre las iteraciones:

### **Gradient Boosting: Una Introducción**

#### **1. ¿Qué es Gradient Boosting?**


Gradient Boosting es un algoritmo de aprendizaje supervisado que combina varios **weak learners** (normalmente árboles de decisión con poca profundidad) en un modelo fuerte y preciso. Es parte de la familia de algoritmos de **boosting**, donde los modelos se entrenan secuencialmente, cada uno corrigiendo los errores del anterior.

Su objetivo principal es **minimizar una función de pérdida**, ajustando progresivamente las predicciones para reducir los errores residuales. Gradient Boosting es ampliamente utilizado tanto para problemas de **regresión** como de **clasificación**.

---

#### **2. ¿Cómo funciona Gradient Boosting?**



El proceso se basa en la idea de ajustar modelos iterativamente para corregir los errores de las predicciones anteriores. El método general sigue estos pasos:

---

##### **Paso 1: Inicialización**
El algoritmo comienza con un modelo base, que suele ser una predicción simple. Por ejemplo:
- En **regresión**, se usa la media de los valores objetivo (\(y_i\)).
- En **clasificación**, se utiliza la probabilidad logarítmica (\(p = 0.5\)).

Esto crea un modelo inicial \(F_0(x)\):
\[
F_0(x) = \text{predicción inicial}
\]

---

##### **Paso 2: Cálculo de los Residuos**
Para cada muestra, se calcula el **residuo**, que representa el error entre el valor real (\(y_i\)) y la predicción actual del modelo (\(F_{t-1}(x_i)\)):
\[
r_i = y_i - F_{t-1}(x_i)
\]
Estos residuos son los valores objetivo que el siguiente weak learner intentará predecir.

---

##### **Paso 3: Entrenamiento del Weak Learner**
Se entrena un **weak learner** (típicamente un árbol de decisión) para predecir los residuos calculados en el paso anterior:
\[
h_t(x) = \text{modelo ajustado a los residuos}
\]

---

##### **Paso 4: Actualización del Modelo**
El modelo se actualiza combinando el modelo previo con una versión ponderada del nuevo weak learner:
\[
F_t(x) = F_{t-1}(x) + \eta \cdot h_t(x)
\]
Donde:
- \(F_t(x)\): Modelo actualizado.
- \(\eta\): Tasa de aprendizaje (learning rate), que controla cuánto contribuye el nuevo weak learner.

---

##### **Paso 5: Iteraciones**
El proceso se repite hasta alcanzar el número de iteraciones especificado o hasta que la función de pérdida deje de mejorar significativamente.

---

#### **3. ¿Por Qué Funciona Gradient Boosting?**

Gradient Boosting minimiza una **función de pérdida** de manera eficiente al aplicar el descenso por gradiente en el espacio de funciones. Esto significa que el algoritmo ajusta cada weak learner para que apunte en la dirección de la mayor reducción del error.

La función de pérdida puede ser:
- **Regresión**: Error cuadrático medio (MSE).
- **Clasificación**: Pérdida logarítmica (log loss) o deviance.

---

**Ejemplo Intuitivo**

1. Supongamos que quieres predecir el precio de una casa.
2. Comienzas con un modelo simple que predice el promedio de los precios de las casas.
3. Luego, calculas cuánto difieren los precios reales de las predicciones iniciales (residuos).
4. Entrenas un modelo (weak learner) para predecir esos residuos.
5. Actualizas el modelo sumando las predicciones del weak learner al modelo actual.
6. Repites el proceso hasta que los residuos sean pequeños.

---

#### **5. Ventajas y Limitaciones de Gradient Boosting**

++Ventajas**

- **Flexibilidad**: Funciona con múltiples funciones de pérdida y puede ajustarse para clasificación y regresión.
- **Precisión**: Suele producir modelos altamente precisos.
- **Robustez**: Es más resistente al sobreajuste cuando se ajustan parámetros como el learning rate.

---

**Limitaciones**

- **Lento en Entrenamiento**: Entrenar modelos secuenciales puede ser más lento que otros métodos como Random Forest.
- **Sensibilidad al Ruido**: Puede sobreajustarse a datos ruidosos si no se configura adecuadamente.
- **Complejidad de Parámetros**: Requiere ajustar parámetros como el número de iteraciones, el learning rate, y la profundidad de los árboles.

---

#### **Resumen Visual**


| **Paso**                    | **Descripción**                                                                 |
|-----------------------------|---------------------------------------------------------------------------------|
| Inicialización              | Comienza con un modelo base \(F_0(x)\).                                        |
| Cálculo de Residuos         | Los residuos representan los errores actuales (\(r_i = y_i - F_{t-1}(x_i)\)).  |
| Weak Learner                | Se entrena un modelo débil para predecir los residuos (\(h_t(x)\)).            |
| Actualización del Modelo    | Se actualiza el modelo sumando una versión ponderada del weak learner (\(\eta \cdot h_t(x)\)). |
| Repetición                  | Se repiten los pasos hasta alcanzar el número de iteraciones deseado.          |

---

#### **Ejemplo en código usando GradientBoostingClassifier de la librería de sklearn para un problema de clasificación**

Fuente: https://scikit-learn.org/dev/modules/generated/sklearn.ensemble.GradientBoostingClassifier.html

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# Generar un conjunto de datos de ejemplo
from sklearn.datasets import make_classification

# Crear un conjunto de datos sintético
X, y = make_classification(
    n_samples=1000,
    n_features=20,
    n_informative=15,
    n_redundant=5,
    random_state=42
)

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Crear y entrenar un modelo de Gradient Boosting
model = GradientBoostingClassifier(
    n_estimators=100,  # Número de árboles (learners débiles son árboles de decisión en esta implementación)
    learning_rate=0.1, # Tasa de aprendizaje --> cuánto aprendizaje de lo obtenido en la iteración incorporo
    max_depth=3,       # Profundidad máxima de los árboles
    random_state=42    # Semilla para reproducibilidad, pues fija el punto de partida
    # Dos parámetros que nos entregan para mover las métricas que vamos a optimizar
    # loss --> la función de pérdida a ser optimizada
    # criterion  --> la función para medir la calidad de splits en los árboles
)

# Entrenar el modelo
model.fit(X_train, y_train)

# Realizar predicciones
y_pred = model.predict(X_test)

# Evaluar el rendimiento del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Exactitud del modelo: {accuracy:.2f}")

# Mostrar un reporte de clasificación
print("Reporte de Clasificación:")
print(classification_report(y_test, y_pred))
```

#### Implementación en código de un Gradiente Boosting manual paso a paso

```python
# Importar bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Crear un conjunto de datos simple
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Una característica, valores entre 0 y 10
y = 5 * X[:, 0] + np.sin(X[:, 0]) * 10 + np.random.randn(100) * 2  # Relación no lineal con ruido

# Visualizar los datos
plt.scatter(X, y, color='blue', label='Datos reales')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Datos para regresión')
plt.legend()
plt.show()

# Configuración inicial
n_iterations = 5
learning_rate = 0.2  # Tasa de aprendizaje
predictions = np.zeros_like(y)  # Predicciones iniciales

# Listas para guardar modelos, errores y predicciones en cada iteración
models = []
errors = []
predictions_per_step = []

# Gradiente Boosting manual paso a paso
for i in range(n_iterations):
    # Calcular los residuos para la iteración anterior (con la excepción de la 1era iteración)
    residuals = y - predictions

    # Ajustar un weak learner (árbol de decisión) a los residuos
    model = DecisionTreeRegressor(max_depth=3, random_state=42)
    # Actualizo el modelo usando los residuos calculados para la iteración anterior
    model.fit(X, residuals)
    models.append(model)

    # Actualizar las predicciones
    update = model.predict(X)
    predictions += learning_rate * update

    # Guardar las predicciones y el error
    predictions_per_step.append(predictions.copy())
    mae = np.mean(abs(y - predictions))
    errors.append(mae)

    # Mostrar información de la iteración
    print(f"Iteración {i + 1}")
    print(f"Error absoluto medio: {mae:.4f}")
    print("-" * 40)

# Visualizar el error a lo largo de las iteraciones
plt.plot(range(1, n_iterations + 1), errors, marker='o', label='Error absoluto medio')
plt.xlabel('Iteración')
plt.ylabel('MAE')
plt.title('Evolución del error')
plt.legend()
plt.show()

# Visualizar las predicciones en cada iteración
for i, step_predictions in enumerate(predictions_per_step, start=1):
    plt.scatter(X, y, color='blue', alpha=0.5, label='Datos reales' if i == 1 else "")
    plt.scatter(X, step_predictions, color='red', alpha=0.5, label=f'Predicciones iteración {i}')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title(f'Iteración {i}: Predicciones vs Datos reales')
    plt.legend()
    plt.show()

# Visualizar las predicciones finales
plt.scatter(X, y, color='blue', label='Datos reales')
plt.scatter(X, predictions, color='green', label='Predicciones finales')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Datos vs Predicciones finales')
plt.legend()
plt.show()
```

## Gradient Boosting vs AdaBoost: Una Comparación



Gradient Boosting y AdaBoost son algoritmos de boosting que combinan múltiples weak learners (normalmente árboles de decisión de baja profundidad) para construir un modelo fuerte y preciso. Aunque tienen similitudes en su objetivo, difieren significativamente en su enfoque y funcionamiento.

---

**1. Similitudes Entre Gradient Boosting y AdaBoost**
- **Weak Learners**: Ambos utilizan modelos débiles, típicamente árboles de decisión con una profundidad pequeña, como base para construir el modelo.
- **Iteraciones Secuenciales**: Los modelos se entrenan de forma iterativa, donde cada modelo intenta corregir los errores del modelo anterior.
- **Reducción del Error**: Ambos algoritmos están diseñados para reducir el error en cada iteración mediante la combinación ponderada de los weak learners.

---

**2. Diferencias Clave**

| **Característica**          | **AdaBoost**                                                                 | **Gradient Boosting**                                                                 |
|-----------------------------|------------------------------------------------------------------------------|--------------------------------------------------------------------------------------|
| **Enfoque Principal**       | Ajusta los pesos de las muestras en función de sus errores de clasificación. | Minimiza una función de pérdida mediante gradientes, ajustando las predicciones residuales. |
| **Tipo de Error**           | Usa un enfoque basado en errores de clasificación.                          | Utiliza funciones de pérdida genéricas, como la pérdida logística para clasificación o la MSE para regresión. |
| **Weak Learners**           | Se entrenan iterativamente con mayor peso en las muestras mal clasificadas. | Los weak learners predicen los residuos (errores) del modelo anterior.              |
| **Combinación de Modelos**  | Pondera los weak learners en función de su precisión (peso \(\alpha_t\)).    | Suma los modelos ponderados por un learning rate para minimizar los residuos.       |
| **Robustez a Ruido**        | Menos robusto al ruido; se ve afectado por datos mal etiquetados.            | Más robusto porque minimiza gradualmente la función de pérdida global.              |
| **Aplicaciones**            | Más común en problemas de clasificación.                                     | Versátil, se aplica tanto en clasificación como en regresión.                       |

---

**3. Funcionamiento de Gradient Boosting**

1. **Inicialización**:
   - Se comienza con un modelo inicial (\(F_0(x)\)), que típicamente predice la media en regresión o las probabilidades en clasificación.

2. **Cálculo de Residuos**:
   - En cada iteración \(t\), se calculan los residuos \(r_i = y_i - F_{t-1}(x_i)\), que representan los errores actuales del modelo.

3. **Ajuste de un Weak Learner**:
   - Se entrena un weak learner (\(h_t(x)\)) para predecir estos residuos.

4. **Actualización del Modelo**:
   - El modelo se actualiza sumando una versión escalada (\(\eta\), learning rate) del weak learner:
     \[
     F_t(x) = F_{t-1}(x) + \eta \cdot h_t(x)
     \]

5. **Iteraciones**:
   - Este proceso se repite, refinando el modelo al reducir los residuos iterativamente.

---

**4. Funcionamiento de AdaBoost**

1. **Inicialización**:
   - Se asignan pesos iguales a todas las muestras: \(w_i = \frac{1}{n}\).

2. **Entrenamiento del Weak Learner**:
   - En cada iteración, se entrena un weak learner usando los pesos actuales.

3. **Cálculo del Peso del Weak Learner**:
   - Se calcula un peso \(\alpha_t\) en función del error ponderado (\(\epsilon_t\)):
     \[
     \alpha_t = \ln\left(\frac{1 - \epsilon_t}{\epsilon_t}\right)
     \]

4. **Actualización de Pesos**:
   - Los pesos de las muestras se actualizan, dando más importancia a las muestras mal clasificadas:
     \[
     w_i \propto w_i \cdot \exp(\alpha_t \cdot \text{Error}(y_i, h_t(x_i)))
     \]

5. **Combinación de Weak Learners**:
   - La predicción final es una combinación ponderada de los weak learners:
     \[
     H(x) = \text{sign}\left(\sum_t \alpha_t \cdot h_t(x)\right)
     \]

---

**5. Ejemplo Visual**
En Gradient Boosting, el modelo ajusta residuos iterativamente:
- **Primera Iteración**: Ajusta el modelo a las etiquetas originales.
- **Iteraciones Posteriores**: Ajusta el modelo para reducir los errores residuales acumulados.

En AdaBoost, el modelo ajusta los pesos de las muestras:
- Las muestras mal clasificadas reciben mayor peso en el siguiente weak learner.
- Esto permite que el modelo se concentre en las áreas problemáticas del conjunto de datos.

---

**6. ¿Cuál Usar?**

| **Escenario**                                | **Recomendación**                                 |
|---------------------------------------------|--------------------------------------------------|
| Clasificación rápida y simple               | Usa **AdaBoost**.                                |
| Problemas con ruido o datos complejos       | Usa **Gradient Boosting**, ya que es más robusto. |
| Control detallado de la función de pérdida  | Usa **Gradient Boosting**.                       |

Gradient Boosting es más flexible y robusto, mientras que AdaBoost es más simple y eficiente en problemas bien definidos.
